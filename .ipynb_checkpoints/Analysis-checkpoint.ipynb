# Predicting Housing Sale Prices in Ames, Iowa

----




### Project Description
With 79 explanatory variables describing (almost) every aspect of residential homes in Ames, Iowa, our goal is to predict the final price of each home of the test data set using the training data set.


### Datasets

80 Columns x 1460 Rows of housing sales data from years 2006 to 2010

Source: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data


```python
# Dependencies and Setup
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Importing Data
data = pd.read_csv("data.csv")
#data.info()
```

```python
# Select non object columns (and Set NA to 0)
# For purposes of this exercise use only numeric variables (treat as continuous)
df = data.select_dtypes(exclude=['object'])
df = df.rename(columns={"1stFlrSF" : "X1stFlrSF", "2ndFlrSF" : "X2ndFlrSF"})
df.fillna(0, inplace=True)
df.isnull().any().any()
df.info()
```

### Initial Modeling and Variable Selection

```python
# Single variable model
import statsmodels.api as sm

i=1 # select indepenedent varaible
model = sm.GLM(df["SalePrice"], df[df.columns[i]]).fit()
print(model.deviance)
print(df.columns[i])
print(model.summary())
```

```python
# Single variable - Calculate Deviance
dev_comp = pd.DataFrame()
fields = []
dev = []

# Calculating Deviance
for i in range(37): 
    fields.append(df.columns[i])
    model = sm.GLM(df["SalePrice"], df[df.columns[i]]).fit()
    dev.append(model.deviance)
        
# Output deviance measures and rank by decending
dev_comp["Fields"] = fields
dev_comp["Deviation"] = dev
dev_comp = dev_comp.sort_values(by="Deviation", ascending=True)
print(dev_comp.head(15))
```

```python
# Model with all numeric varibles (continuous treatment)
y = df["SalePrice"]
cols = [
 'LotArea',
 'OverallQual',
 'OverallCond',
 'YearBuilt',
 'TotalBsmtSF',
 '1stFlrSF',
 'GrLivArea',
 'FullBath',
 'GarageCars',
 'WoodDeckSF'
]
X = data[cols].copy()
corr = X.corr()
corr.style.background_gradient(cmap='coolwarm',axis=None).set_precision(2)
```

```python
# Select fields with P value less than 0.05
X = (X - X.min()) / (X.max() - X.min())
X = sm.add_constant(X)

model = sm.GLM(y, X, family=sm.families.Gaussian()).fit()
model.summary()
```

### Model Setup

```python
# split into traingin and testing models
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

### Models

```python
# Identiy link function
ind = sm.GLM(y_train, X_train, family=sm.families.Gaussian()).fit()
print(ind.summary())

print("")
print("")

# Identiy log function
log = sm.GLM(y_train, X_train, family=sm.families.Gaussian(sm.families.links.log)).fit()
print(log.summary())
```

```python
# Plots for Train Dataset
min = 0
max = 800000
pred_ind = ind.predict(X_train)
pred_log = log.predict(X_train)

fig = plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Actual vs Predicted - Train Datset")
plt.scatter(pred_ind, y_train, s = 5, color="blue", label="Ind")
plt.scatter(pred_log, y_train, s = 5, color="red", label="Log")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.plot([0,max],[0,max])
plt.xticks(rotation = 60)
plt.xlim(min,max)
plt.ylim(min,max)

plt.subplot(1,2,2)
plt.title("Q-Q Plot - Train Datset")
plt.scatter(np.sort(pred_ind), np.sort(y_train), s = 5, color="blue", label="Ind")
plt.scatter(np.sort(pred_log), np.sort(y_train), s = 5, color="red", label="Log")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.plot([0,max],[0,max])
plt.xticks(rotation = 60)
plt.xlim(min,max)
plt.ylim(min,max)

plt.show()
```

```python
# Plots for Test Dataset
min = 0
max = 800000
pred_ind_test = ind.predict(X_test)
pred_log_test = log.predict(X_test)

fig = plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Actual vs Predicted - Test Datset")
plt.scatter(pred_ind_test, y_test, s = 5, color="blue", label="Ind")
plt.scatter(pred_log_test, y_test, s = 5, color="red", label="Log")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.plot([0,max],[0,max])
plt.xticks(rotation = 60)
plt.xlim(min,max)
plt.ylim(min,max)

plt.subplot(1,2,2)
plt.title("Q-Q Plot - Test Datset")
plt.scatter(np.sort(pred_ind_test), np.sort(y_test), s = 5, color="blue", label="Ind")
plt.scatter(np.sort(pred_log_test), np.sort(y_test), s = 5, color="red", label="Log")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.plot([0,max],[0,max])
plt.xticks(rotation = 60)
plt.xlim(min,max)
plt.ylim(min,max)

plt.show()
```

```python
# Calculate the mean squared error
from sklearn.metrics import mean_squared_error
mse = np.sqrt(mean_squared_error(y_train,pred_log))
avg = np.mean(y_train)
print("Train Data - MSE: " + '{0:,.0f}'.format(mse) + " and Mean Value: " + '{0:,.0f}'.format(avg))

mse = np.sqrt(mean_squared_error(y_test,pred_log_test))
avg = np.mean(y_test)
print("Test Data - MSE: " + '{0:,.0f}'.format(mse) + " and Mean Value: " + '{0:,.0f}'.format(avg))
```
